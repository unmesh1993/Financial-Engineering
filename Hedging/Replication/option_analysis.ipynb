```python
class OptionAnalysis:
    def __init__(self):
        pass

    def plot_pricing_error_vs_maturity(self, records):
        # Set a bright and clean theme
        sns.set_theme(style="whitegrid", context="talk", font_scale=1)
        plt.rcParams.update({'axes.facecolor': 'whitesmoke'})  # Lighter background for the plot

        org_records = records.copy()
        org_records.sort_values(by='Maturity', inplace=True)
        org_records['diff'] = org_records['cal_price'] - org_records['obs_price']

        calls = org_records[org_records['option_type'] == 'call']
        puts = org_records[org_records['option_type'] == 'put']

        fig, axes = plt.subplots(1, 2, figsize=(14, 7), sharey=True)

        for ax, data, title, color in zip(
            axes,
            [calls, puts],
            ['Call Options', 'Put Options'],
            ['cornflowerblue', 'mediumseagreen']  # Brighter, softer colors
        ):
            sns.scatterplot(
                data=data,
                x='Maturity',
                y='diff',
                color=color,
                s=130,
                edgecolor='white',  # White edges for a cleaner look
                linewidth=1,
                ax=ax
            )

            # Annotate with integer strike values
            for _, row in data.iterrows():
                label = f"{int(round(row['strike']))}"
                ax.annotate(label,
                            (row['Maturity'], row['diff']),
                            textcoords="offset points", xytext=(0, 10),
                            ha='center', fontsize=10, color='black')  # Black labels for contrast

            ax.axhline(0, color='gray', linestyle='--', linewidth=1)
            ax.set_title(title, fontsize=16, fontweight='bold')
            ax.set_xlabel("Maturity (Years)", fontsize=12)
            ax.set_ylabel("Pricing Error (Model - Market)", fontsize=12)
            ax.tick_params(axis='both', which='major', labelsize=12)
            ax.grid(True, linestyle=':', linewidth=0.8, color='gray')  # Lighter grid lines

        plt.tight_layout()
        plt.ylim(-10, 10)
        plt.suptitle("Option Pricing Error vs Maturity", fontsize=18, fontweight='bold')
        plt.subplots_adjust(top=0.88)
        plt.show()
```
